```python
import matplotlib.pyplot as plt

# Sample data (counts of males and females)
genders = ['Male', 'Female']
counts = [45, 55]  # Example counts

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(genders, counts, color=['blue', 'pink'])  # Adjust colors as desired
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Gender in a Population')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```
